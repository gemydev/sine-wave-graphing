```python
import numpy as np
import matplotlib.pyplot as plt
```

The signal is written as <br>
**`s(t) = A sin(2πft + ϕ)`** <br>

**A** is called the **amplitude** of the wave, i.e. the largest value of the wave above or below the *horizontal axis*.<br>
**ϕ** is the **phase** of the wave, which means how far the wave is shifted to the left or the right.<br>
**t** is the **time** between successive peaks<br>
**f** is the **frequency** of the wave<br>
**π** is equall 3.14 or 22/7

```python
freq = float(input("Enter Freuancy : "))
amplitude = float(input("Enter Amplitude : "))
phase = float(input("Enter Phase : "))
input_time = float(input("Enter Time : "))
```

```python
# generate a list of time points on equal distances
time = np.linspace(0,input_time,int(input_time*100))
```

```python
# generate a list of y signals points based on the general rule
y = amplitude * np.sin(2*np.pi*freq*time + phase)
```

```python
# ploting this sine wave by represent time , y in plot graph
plt.figure(figsize = (15, 15))
plt.plot(time, y, 'b')
plt.ylabel('Amplitude')
plt.xlabel('Time (s)')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.show()
```

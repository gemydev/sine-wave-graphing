```python
import numpy as np
import matplotlib.pyplot as plot
```

```python
amplitude = 2
frequancy = 50
phase = 0
cycle = 1
T = 1 / frequancy
Fs = 100 * frequancy
Ts = 1 / Fs
```

```python
time = np.arange(0, T * cycle, Ts)
```

```python
y = amplitude * np.sin(2 * np.pi * frequancy * time + phase)
```

```python
plot.plot(y)

# Give a title for the sine wave plot
plot.title('Sine wave')

# Give x axis label for the sine wave plot
plot.xlabel('Time')

# Give y axis label for the sine wave plot
plot.ylabel('Amplitude')
```
